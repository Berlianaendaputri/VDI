```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('/content/dataset disney.csv')
df.head()

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='role;', data=df)
plt.title('Jumlah Aktor per Peran')
plt.xlabel('Peran')
plt.ylabel('Jumlah')
plt.show()
```

Visualisasi diatas menggunakan grafik batang yang menunjukkan sebaran jumlah aktor berdasarkan peran yang dimainkannya.
sebaran pertama yang menunjukkan jumlah nya mecapai 25000
dan sebaran kedua yaitu director peran mengacu pada aktor yang juga menyutradarai, namun jumlahnya jauh lebih kecil, sekitar 1.000.
Pada visualisasi tersebut dapat terlihat bahwa ketinggian batang terlihat jelas dan  perbedaan skala yang jelas antara kedua peran tersebut bisa langsung terlihat.Namun terdapat kesalahan kecil pada label "ACTOR;" pada sumbu X.
muncul dua kali sehingga dapat membingungkan pembaca. Selain itu, penggunaan warna yang sama untuk semua batang menyulitkan untuk segera membedakan kategori.
Namun visualisasi ini secara efektif menunjukkan bahwa mayoritas orang di industri hanya berperan sebagai aktor, dan sangat sedikit yang juga menjadi sutradara.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='person_id', y='role;', data=df)
plt.title('Hubungan Antara Person ID dan role;')
plt.xlabel('Person ID')
plt.ylabel('role;')
plt.show()
```

```python
df_char_count = df.groupby('person_id')['character'].nunique().reset_index(name='jumlah_karakter')
plt.figure(figsize=(10, 6))
sns.barplot(x='person_id', y='jumlah_karakter', data=df_char_count)
plt.title('Jumlah Karakter yang Diperankan per Person')
plt.xlabel('Person ID')
plt.ylabel('Jumlah Karakter')
plt.show()
```

Visualisasi diatas menggunakan grafik batang yang menunjukkan jumlah karakter yang dimainkan oleh setiap orang dalam kumpulan data.Sumbu X melambangkan "ID orang" dan sumbu Y melambangkan "jumlah karakter".Dari visualisasi tersebut dapat dilihat bahwa kebanyakan orang hanya memainkan beberapa karakter saja.
Hal tersebut dapat terlihat dari banyaknya bar pendek yang  mendekati angka 0 pada sumbu Y.Namun, ada beberapa orang memainkan lebih banyak karakter. ditampilkan di beberapa bar besar, dengan beberapa orang memainkan lebih dari 40 karakter pada puncaknya.sehingga menunjukkan bahwa meskipun sebagian besar orang dalam kumpulan data hanya memainkan beberapa karakter,  beberapa orang sangat produktif dan memainkan banyak karakter pada saat yang bersamaan.
Selain itu, bagian bawah grafisnya, khususnya Visualisasi ini menunjukkan perbedaan jumlah karakter yang dimainkan, kebanyakan orang  hanya memainkan beberapa karakter, sedangkan beberapa orang memainkan banyak karakter.



```python
from wordcloud import WordCloud

text = ' '.join(df['character'].dropna().astype(str).tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Visualisasi diatas merupakan word cloud atau awan, kata yang di tampilkan merupakan  kata-kata yang paling sering muncul dalam  kumpulan data.
Dalam visualisasi pada dataset disney kata "voice" dan "Additional"  paling sering muncul yang artinya menunjukkan bahwa kata-kata tersebut paling sering muncul dalam data yang dianalisis.tetapi  kata-kata lain seperti uncredited,self, dan voice juga  terlihat, dimana istilah-istilah tersebut juga sering digunakan, namun lebih jarang dibandingkan voice dan Additional.
dimana kata pada visualisasi data tersebut dapat dilihat bahwa yang paling sering muncul itu merupakan data yang paling sering digunakan dalam pembuatan film pada disney.Kehadiran variasi kata lain seperti "muppet", "penari", "narator", "penyiar", dan "pemain" menunjukkan bahwa data ini mungkin merujuk pada peran yang dimainkan oleh individu dalam proyek yang berbeda.
